# Super Car Spotter

## Introduction/Business Problem 
This project hopes to show tourists/sightseers where they are most likely to spot supercars in three major US cities; Chicago, Austin, and Phoenix. 

Looking for supercars when touring or being out in a new city can be a hastle if you don't know where to look. This car spotting project hopes to solve this ***problem*** and can be something that allows the everyday American to easily visualize any luxury car goals they may have. 

## Data Description
Much like bird watching, we will take into account a number of geographical and socioeconomic factors to figure out where supercars are more likely to be. Data for each of the three cities being considered will be pulled from the SquareSpace API.

The Data used will include restaurants, recreational activities, housing prices, and high end shopping centers. A streets proximity to these factors will give it a score that indicates your likelyhood of spotting a super car in an hour long window during the day.
